```python
import pandas as pd
import requests

# Obtener los datos de la API
api_key = "#########################"  # Tu clave API
city = "New York"
url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}"

response = requests.get(url)
data = response.json()

# Normalizar los datos de JSON a un Dataframe
df = pd.json_normalize(data)

print(df)
```

```python
df
```

```python
# Función para convertir Unix timestamp a formato legible
from datetime import datetime, timezone
def convert_unix_to_datetime(unix_timestamp):
    return datetime.fromtimestamp(unix_timestamp, tz=timezone.utc).strftime('%d-%m-%Y %H:%M:%S')
```

```python
# Aplicar la función a las columnas relevantes del DataFrame
df['dt'] = df['dt'].apply(convert_unix_to_datetime)
df['sys.sunrise'] = df['sys.sunrise'].apply(convert_unix_to_datetime)
df['sys.sunset'] = df['sys.sunset'].apply(convert_unix_to_datetime)
```

```python
df
```

```python
# Desanidar la columna 'weather'
weather_df = pd.json_normalize(data['weather'])
weather_df.columns = ['weather.' + col for col in weather_df.columns]  # Renombrar las columnas

# Concatenar el DataFrame original con el DataFrame desanidado
df = pd.concat([df.drop(columns=['weather']), weather_df], axis=1)
```
