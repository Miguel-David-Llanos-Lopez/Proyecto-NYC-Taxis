```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
Light_Duty_Vehicles = pd.read_csv(r'Data\Light Duty Vehicles.csv')
Light_Duty_Vehicles.head(5)
```

```python
Light_Duty_Vehicles.tail(5)
```

```python
# Información de las columnas
print(Light_Duty_Vehicles.columns)
```

```python
# Información general del dataset
print(Light_Duty_Vehicles.info())
```

```python
# Descripción estadística de las columnas numéricas
print(Light_Duty_Vehicles.describe())
```

```python
# Descripción estadística de las columnas categóricas

print(Light_Duty_Vehicles.describe(include=['object']))
```

```python
# Verificar el total de valores nulos por columna
print(Light_Duty_Vehicles.isnull().sum())
```

```python
# Verificar el porcentaje de valores nulos por columna
print((Light_Duty_Vehicles.isnull().sum() / len(Light_Duty_Vehicles)) * 100)
```

```python
# Verificar si hay duplicados
print(Light_Duty_Vehicles.duplicated().sum())
```

```python
# Crear un histograma para mostrar la distribución de vehículos por año de modelo
plt.figure(figsize=(10, 6))
sns.histplot(data=Light_Duty_Vehicles, x='Model Year', bins=30, kde=False)
plt.title('Distribución de Vehículos por Año')
plt.xlabel('Año del Modelo')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Crear un pie chart para mostrar la distribución de categorías de vehículos
plt.figure(figsize=(8, 8))
Light_Duty_Vehicles['Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribución de Categorías de Vehículos')
plt.ylabel('')  # Eliminar etiqueta del eje y para un mejor aspecto visual
plt.show()
```

```python
# Crear un gráfico de barras para mostrar la distribución de tipos de combustible
plt.figure(figsize=(10, 6))
sns.countplot(data=Light_Duty_Vehicles, x='Fuel', order=Light_Duty_Vehicles['Fuel'].value_counts().index)
plt.title('Distribución de Tipos de Combustible')
plt.xlabel('Tipo de Combustible')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mejor legibilidad
plt.show()
```

```python
# Crear un boxplot para mostrar la distribución de la autonomía en modo eléctrico por categoría de vehículo
plt.figure(figsize=(12, 6))
sns.boxplot(data=Light_Duty_Vehicles, x='Category', y='Electric-Only Range')
plt.title('Autonomía en Modo Eléctrico por Categoría de Vehículo')
plt.xlabel('Categoría de Vehículo')
plt.ylabel('Autonomía en Modo Eléctrico (millas)')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mejor legibilidad
plt.show()
```

```python
# Crear un scatter plot para visualizar la relación entre el año del modelo y la autonomía en modo eléctrico
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Light_Duty_Vehicles, x='Model Year', y='Electric-Only Range')
plt.title('Relación entre Año del Modelo y Autonomía Eléctrica')
plt.xlabel('Año del Modelo')
plt.ylabel('Autonomía en Modo Eléctrico (millas)')
plt.show()
```

```python
# Calcular la matriz de correlación entre las variables numéricas clave
correlation_matrix = Light_Duty_Vehicles[['Model Year', 'Electric-Only Range']].corr()

# Crear un heatmap para mostrar la correlación entre variables
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlación entre Variables Clave')
plt.show()
```
