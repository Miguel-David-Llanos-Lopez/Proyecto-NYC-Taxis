```python
plt.figure(figsize=(10, 6))
sns.histplot(data=Vehicle_Fuel_Economy, x='Year', bins=30, kde=False)
plt.title('Distribución de Vehículos por Año')
plt.xlabel('Año del Modelo')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Boxplot para mostrar la distribución de emisiones de CO2 por clase de vehículo
plt.figure(figsize=(30, 8))
sns.boxplot(x='VClass', y='co2', data=Vehicle_Fuel_Economy)
plt.title('Emisiones de CO2 por Clase de Vehículo')
plt.xlabel('Clase de Vehículo')
plt.ylabel('Emisiones de CO2 (gramos por milla)')
plt.xticks(rotation=80)
plt.show()
```

```python
# Scatter plot para mostrar la relación entre emisiones y eficiencia de combustible
plt.figure(figsize=(10, 6))
sns.scatterplot(x='co2', y='feScore', data=Vehicle_Fuel_Economy)
plt.title('Emisiones de CO2 vs. Puntaje de Eficiencia de Combustible')
plt.xlabel('Emisiones de CO2 (gramos por milla)')
plt.ylabel('Puntaje de Eficiencia de Combustible')
plt.show()
```

```python
# Agrupar los datos por año y calcular el costo anual promedio de combustible
grouped_data = Vehicle_Fuel_Economy.groupby('Year')['fuelCost08'].mean().reset_index()

# Histograma: Distribución de Costos de Combustible Anuales por Año
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='fuelCost08', data=grouped_data, palette='viridis')
plt.title('Costo Anual Promedio de Combustible por Año')
plt.xlabel('Año')
plt.ylabel('Costo Anual de Combustible ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtrar el DataFrame para años a partir de 2014
df_vehicle_filtered = Vehicle_Fuel_Economy[Vehicle_Fuel_Economy['Year'] >= 2014]

# Calcular el IQR (rango intercuartílico) de la columna 'co2'
Q1 = df_vehicle_filtered['co2'].quantile(0.25)
Q3 = df_vehicle_filtered['co2'].quantile(0.75)
IQR = Q3 - Q1

# Definir límites para considerar valores atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los valores atípicos en la columna 'co2'
df_vehicle_filtered = df_vehicle_filtered[(df_vehicle_filtered['co2'] >= lower_bound) & (df_vehicle_filtered['co2'] <= upper_bound)]

# Crear el gráfico con los datos filtrados
plt.figure(figsize=(10, 6))
sns.regplot(x='Year', y='co2', data=df_vehicle_filtered, scatter_kws={'alpha': 0.3}, line_kws={"color": "r", "alpha": 0.7, "lw": 2})
plt.title('Año de Fabricación vs. Emisiones de CO2 (Desde 2010) - Sin Valores Atípicos')
plt.xlabel('Año de Fabricación')
plt.ylabel('Emisiones de CO2 (gramos/milla)')
plt.xlim(2013,2025)  # Asegurarse de que el eje x comience en 2014
plt.show()
```

```python
# Boxplot para mostrar la distribución de costos de combustible por tipo de combustible
plt.figure(figsize=(10, 6))
sns.boxplot(data=Vehicle_Fuel_Economy, x='fuelType1', y='fuelCost08')
plt.xlabel('Tipo de Combustible')
plt.ylabel('Costo de Combustible')
plt.title('Costos de Combustible por Tipo de Combustible')
plt.show()
```

```python
# Filtrar las columnas relevantes
columns_to_consider = ["Manufacturer", "Model", "fuelCost08", "feScore"]
filtered_data = Vehicle_Fuel_Economy[columns_to_consider]

# Eliminar filas con valores faltantes
filtered_data_cleaned = filtered_data.dropna()

# Identificar los vehículos más eficientes
top_efficient_vehicles = filtered_data_cleaned.nlargest(10, 'feScore')

# Calcular los costos anuales de combustible para los vehículos más eficientes
top_efficient_costs = top_efficient_vehicles[['Manufacturer', 'Model', 'fuelCost08', 'feScore']]

# Crear un gráfico de barras para mostrar los costos anuales de los vehículos más eficientes
plt.figure(figsize=(12, 6))
sns.barplot(x='feScore', y='fuelCost08', hue='Manufacturer', data=top_efficient_costs)
plt.xlabel('Puntaje de Eficiencia de Combustible')
plt.ylabel('Costo Anual de Combustible ($)')
plt.title('Costos Anuales de Combustible de los Vehículos Más Eficientes')
plt.legend(title='Fabricante', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Heatmap: Correlación entre Autonomía y Eficiencia
correlation_matrix = Vehicle_Fuel_Economy[['highwayE', 'range', 'feScore', 'co2']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlación entre Autonomía y Eficiencia de Combustible')
plt.show()
```

```python
# Calculando la media de emisiones de CO2 por marca
top_co2_brands = Vehicle_Fuel_Economy.groupby('Manufacturer')['co2'].mean().sort_values(ascending=False).head(10)

# Graficando las marcas con mayores emisiones
plt.figure(figsize=(10, 6))
sns.barplot(x=top_co2_brands.values, y=top_co2_brands.index)
plt.xlabel('promedio de CO2 (grams/mile)')
plt.ylabel('Fabricante')
plt.title('Top 10 Fabricantes con las Emisiones de CO2 más Altas')
plt.show()
```

```python
# Calculando la eficiencia promedio en ciudad por marca
top_fuel_efficiency_brands = Vehicle_Fuel_Economy.groupby('Manufacturer')['city08'].mean().sort_values(ascending=False).head(10)

# Graficando las marcas con mayor eficiencia de combustible
plt.figure(figsize=(10, 6))
sns.barplot(x=top_fuel_efficiency_brands.values, y=top_fuel_efficiency_brands.index)
plt.xlabel('Promedio de MPG en Ciudad')
plt.ylabel('Fabricante')
plt.title('Top 10 Fabricantes con Mayor Eficiencia de Combustible (MPG en Ciudad)')
plt.show()
```

```python
# Tendencia del Costo Anual de Combustible a lo Largo de los Años
plt.figure(figsize=(10, 6))
sns.lineplot(data=Vehicle_Fuel_Economy, x='Year', y='fuelCost08')
plt.xlabel('Año')
plt.ylabel('Costo Anual de Combustible')
plt.title('Tendencia del Costo Anual de Combustible a lo Largo de los Años')
plt.show()
```

```python
# Relación entre la eficiencia eléctrica y la autonomía
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Vehicle_Fuel_Economy, x='cityE', y='range', alpha=0.5)
plt.xlabel('Eficiencia Eléctrica (MPGe)')
plt.ylabel('Rango')
plt.title('Relación entre la Eficiencia Eléctrica y la Autonomía')
plt.show()
```
