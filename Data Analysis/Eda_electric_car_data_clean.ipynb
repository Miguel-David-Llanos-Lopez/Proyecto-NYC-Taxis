```python
# Histograma: Distribución de Vehículos por Marca y Modelo
plt.figure(figsize=(30, 6))
sns.histplot(data=ElectricCarData_Clean, x='Brand', kde=False)
plt.title('Distribución de Vehículos por Marca')
plt.xlabel('Marca del Vehículo')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Scatter Plot: Relación entre Velocidad Máxima y Autonomía
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TopSpeed_KmH', y='Range_Km', data=ElectricCarData_Clean)
plt.title('Relación entre Velocidad Máxima y Autonomía')
plt.xlabel('Velocidad Máxima (km/h)')
plt.ylabel('Autonomía (km)')
plt.show()
```

```python
# Boxplot: Distribución del Consumo de Energía
plt.figure(figsize=(10, 6))
sns.boxplot(y='Efficiency_WhKm', data=ElectricCarData_Clean)
plt.title('Distribución del Consumo de Energía')
plt.ylabel('Consumo de Energía (Wh/km)')
plt.show()
```

```python
# Histograma: Distribución de Precios de Vehículos Eléctricos
plt.figure(figsize=(10, 6))
sns.histplot(data=ElectricCarData_Clean, x='PriceEuro', kde=True)
plt.title('Distribución de Precios de Vehículos Eléctricos')
plt.xlabel('Precio (Euros)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Convertir columnas relevantes a valores numéricos si es necesario
ElectricCarData_Clean['TopSpeed_KmH'] = pd.to_numeric(ElectricCarData_Clean['TopSpeed_KmH'], errors='coerce')
ElectricCarData_Clean['Range_Km'] = pd.to_numeric(ElectricCarData_Clean['Range_Km'], errors='coerce')
ElectricCarData_Clean['Efficiency_WhKm'] = pd.to_numeric(ElectricCarData_Clean['Efficiency_WhKm'], errors='coerce')
ElectricCarData_Clean['FastCharge_KmH'] = pd.to_numeric(ElectricCarData_Clean['FastCharge_KmH'], errors='coerce')
ElectricCarData_Clean['PriceEuro'] = pd.to_numeric(ElectricCarData_Clean['PriceEuro'], errors='coerce')
ElectricCarData_Clean['Seats'] = pd.to_numeric(ElectricCarData_Clean['Seats'], errors='coerce')
ElectricCarData_Clean['AccelSec'] = pd.to_numeric(ElectricCarData_Clean['AccelSec'], errors='coerce')


# Seleccionar columnas relevantes para el heatmap
columns_for_heatmap = ['TopSpeed_KmH', 'Range_Km', 'Efficiency_WhKm', 'FastCharge_KmH', 'PriceEuro', 'Seats', 'AccelSec']

# Eliminar filas con valores nulos en las columnas seleccionadas
data_cleaned = ElectricCarData_Clean[columns_for_heatmap].dropna()

# Calcular la matriz de correlación
correlation_matrix = data_cleaned.corr()

# Crear el heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap de Correlación entre Variables Clave de Vehículos Eléctricos')
plt.show()
```
