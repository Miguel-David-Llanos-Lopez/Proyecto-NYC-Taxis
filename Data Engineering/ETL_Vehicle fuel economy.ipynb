```python
import pandas as pd
```

```python
data = pd.read_csv("../Data/Vehicle Fuel Economy Data.csv")
```

```python
display(data)
```

```python
columns_to_keep = ['Year','Manufacturer','Model', 'charge240','feScore', 'fuelCost08', 'fuelType1', 'VClass']
```

```python
df_clean = data[columns_to_keep]
```

```python
df_clean
```

```python
Clases = ['Subcompact Cars','Compact Cars','Minicompact Cars']

# Filtrar el DataFrame por los modelos necesarios
filtered_df = df_clean[df_clean['VClass'].isin(Clases)]
filtered_df = filtered_df[filtered_df['Year'] > 2015]
```

```python
filtered_df
```

```python
combustibles = filtered_df[['fuelType1']]
combustibles = combustibles.drop_duplicates().reset_index().drop(columns='index')
modelos = filtered_df[['Model']]
modelos = modelos.drop_duplicates().reset_index().drop(columns='index')
clase = filtered_df[['VClass']]
clase = clase.drop_duplicates().reset_index().drop(columns='index')
```

```python

def create_mapping(df_column):
    return {value: index for index, value in df_column.items()}

combustibles_mapping = create_mapping(combustibles['fuelType1'])
modelos_mapping = create_mapping(modelos['Model'])
clase_mapping = create_mapping(clase['VClass'])

def replace_with_id(df_column, mapping):
    return df_column.map(mapping)

filtered_df['fuelType1'] = replace_with_id(filtered_df['fuelType1'], combustibles_mapping)
filtered_df['Model'] = replace_with_id(filtered_df['Model'], modelos_mapping)
filtered_df['VClass'] = replace_with_id(filtered_df['VClass'], clase_mapping)
```

```python
combustibles.to_csv('../Data/Data Prosesada/combustibles de vehiculos.csv')
modelos.to_csv('../Data/Data Prosesada/modelos de vehiculos.csv')
clase.to_csv('../Data/Data Prosesada/clases de vehiculos.csv')
filtered_df.to_csv('../Data/Data Prosesada/vehiculos.csv')
```
